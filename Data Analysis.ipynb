```python
# average distance traveled
# 22.43 was taken from https://www.cyclingweekly.com/news/latest-news/british-cyclists-fastest-in-world-strava-303384#0Y6FAtO7Cx4JgP2B.99
def calculate_distance(row):
    distance = 22.43 * row["Duration"] / 60 / 60
    if row["Trip Route Category"] == "One Way":
        # radius of earth
        R = 6373.0
        
        lat1 = radians(row["Starting Station Latitude"])
        lon1 = radians(row["Starting Station Longitude"])
        lat2 = radians(row["Ending Station Latitude"])
        lon2 = radians(row["Ending Station Longitude"])
        
        dlon = lon2 - lon1
        dlat = lat2 - lat1
        
        # Haversine formula
        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))

        distance = R * c
    return distance

data["Distance"] = data.apply(calculate_distance, axis=1)

data["Distance"].mean()
```

```python
# combinations of passholder types and trip route category
df = data.drop(["Start Time", "End Time", "Plan Duration", "Distance", "Duration", "Trip ID", "Starting Station ID", "Starting Station Latitude", "Starting Station Longitude", "Ending Station ID", "Ending Station Latitude","Ending Station Longitude", "Bike ID"], axis=1)
df.groupby(['Trip Route Category','Passholder Type']).size().reset_index().rename(columns={0:'count'})
```

```python
# graph depicting running average duration by passholder type
pd.options.mode.chained_assignment = None 

df1 = data[["Passholder Type", "Duration", "Start Time"]]
flex_df = df1[df1["Passholder Type"] == "Flex Pass"]
flex_df['mean'] = flex_df['Duration'].rolling(1000).mean()
monthly_df = df1[df1["Passholder Type"] == "Monthly Pass"]
monthly_df['mean'] = monthly_df['Duration'].rolling(1000).mean()
walk_df = df1[df1["Passholder Type"] == "Walk-up"]
walk_df['mean'] = walk_df['Duration'].rolling(1000).mean()

plt.plot('Start Time', 'mean', data=flex_df, color='blue', label="flex")
plt.plot('Start Time', 'mean', data=monthly_df, color='red', label="monthly")
plt.plot('Start Time', 'mean', data=walk_df, color='green', label="walk-up")

total_df = pd.concat([flex_df, monthly_df, walk_df])
out = total_df.to_json(orient='records')[1:-1].replace('},{', '} {')

with open('passholder_vs_duration.json', 'w') as f:
    f.write(out)

total_df
# graph shows that the walk-up riders have a significantly higher duration than the flex/monthly pass users and
# more variation as well
```

```python
# graph depicting one-way vs. round trip for duration
pd.options.mode.chained_assignment = None 

df2 = data[["Trip Route Category", "Duration", "Start Time"]]
round_df = df2[df2["Trip Route Category"] == "Round Trip"]
round_df['mean'] = round_df['Duration'].rolling(1000).mean()
one_way_df = df2[df2["Trip Route Category"] == "One Way"]
one_way_df['mean'] = one_way_df['Duration'].rolling(1000).mean()

plt.plot('Start Time', 'mean', data=round_df, color='blue', label="round")
plt.plot('Start Time', 'mean', data=one_way_df, color='red', label="one")

total_df = pd.concat([round_df, one_way_df])

out = total_df.to_json(orient='records')[1:-1].replace('},{', '} {')

with open('trip_category_vs_duration.json', 'w') as f:
    f.write(out)

# graph shows that there's significantly more variation in the one way trips than there is in the round trips.
# it's expected that the round trips would be longer than the one-way trips, but the variation is interesting.
```
